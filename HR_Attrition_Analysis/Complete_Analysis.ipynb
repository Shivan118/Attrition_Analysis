## Problem Statement:

As a Data Analyst in our HR firm, we aim to analyze and understand the factors influencing employee attrition within our organization. The dataset provided contains various features related to employees, their demographics, job roles, and satisfaction levels. 

The objective is to identify patterns and insights that can help us predict and mitigate attrition rates effectively.


## Key Features:

**Attrition:** Whether the employee has left the company (Yes/No).


**Business Travel:** Frequency of employee's business travel (Non-Travel, Travel_Rarely, Travel_Frequently).


**CF_age band:** Categorical feature representing the age group of employees.

**CF_attrition label:** A categorical label for attrition.

**Department:** Department where the employee works.

**Education Field:** Field of education of the employee.

**emp no:** Employee number.

**Employee Number:** Employee number (similar to emp no).

**Gender:** Gender of the employee.

**Job Role:** Role or position of the employee within the organization.

**Marital Status:** Marital status of the employee.

**Over Time:** Whether the employee works overtime (Yes/No).

**Over18:** Whether the employee is over 18 years old (Yes/No).

**Training Times Last Year:** Number of training sessions attended by the employee last year.

**-2:** Not specified (considered as an unidentified feature).

**0:** Not specified (considered as an unidentified feature).

**Age:** Age of the employee.

**CF_current Employee:** Categorical label indicating if the employee is current.

**Daily Rate:** Daily rate of pay for the employee.

**Distance From Home:** Distance of employee's residence from the workplace.

**Education:** Level of education attained by the employee (1: 'Below College', 2: 'College', 3: 'Bachelor', 4: 'Master', 5: 'Doctor').

**Employee Count:** Number of employees (possibly redundant).

**Environment Satisfaction:** Satisfaction level with the work environment (1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High').

**Hourly Rate:** Hourly rate of pay for the employee.

**Job Involvement:** Level of involvement in the job (1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High').

**Job Level:** Level of the job (1 to 5).

**Job Satisfaction:** Satisfaction level with the job (1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High').

**Monthly Income:** Monthly income of the employee.

**Monthly Rate:** Monthly rate of pay for the employee.

**Num Companies Worked:** Number of companies the employee has worked for.

**Percent Salary Hike:** Percentage increase in salary.

**Performance Rating:** Performance rating of the employee (1 to 4).

**Relationship Satisfaction:** Satisfaction level with work relationships (1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High').

**Standard Hours:** Standard hours of work for the employee.

**Stock Option Level:** Level of stock options for the employee (0 to 3).

**Total Working Years:** Total years of work experience.

**Work Life Balance:** Satisfaction level with work-life balance (1: 'Bad', 2: 'Good', 3: 'Better', 4: 'Best').

**Years At Company:** Years spent at the current company.

**Years In Current Role:** Years spent in the current role.

**Years Since Last Promotion:** Years since the last promotion.

**Years With Curr Manager:** Years spent with the current manager.


```python
# Importing Pandas, Numpy, Matplotlib, Seaborn and Warings Library.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import the CSV Data as Pandas DataFrame
df = pd.read_excel("HR Data.xlsx")
```

```python
# Show all columns
df.columns
```

```python
# SHow the complete information
df.info()
```

```python
# Summary statistics for numerical columns
df.describe()  
```

```python
# Attrition Rate Analysis
attrition_counts = df['Attrition'].value_counts()
attrition_rate = attrition_counts['Yes'] / len(df) * 100
print(f"Attrition Rate: {attrition_rate:.2f}%")
```

```python
# Attrition by Demographic Features
demographic_features = ['Business Travel', 'Gender', 'Marital Status', 'Department', 'Education Field']
for feature in demographic_features:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=feature, hue='Attrition')
    plt.title(f'Attrition by {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.show()
```

```python
# Attrition by Job-related Factors
job_related_factors = ['Job Role', 'Job Level', 'Job Satisfaction', 'Performance Rating']
for factor in job_related_factors:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=factor, hue='Attrition')
    plt.title(f'Attrition by {factor}')
    plt.xlabel(factor)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Attrition and Work Environment
work_environment_factors = ['Environment Satisfaction', 'Work Life Balance']
for factor in work_environment_factors:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=factor, hue='Attrition')
    plt.title(f'Attrition by {factor}')
    plt.xlabel(factor)
    plt.ylabel('Count')
    plt.show()
```

```python
# Attrition and Performance
performance_factors = ['Percent Salary Hike', 'Performance Rating']
for factor in performance_factors:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df, x='Attrition', y=factor)
    plt.title(f'Attrition vs {factor}')
    plt.xlabel('Attrition')
    plt.ylabel(factor)
    plt.show()
```

```python
# Attrition and Tenure
tenure_factors = ['Years At Company', 'Years In Current Role', 'Years Since Last Promotion', 'Years With Curr Manager']
for factor in tenure_factors:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df, x='Attrition', y=factor)
    plt.title(f'Attrition vs {factor}')
    plt.xlabel('Attrition')
    plt.ylabel(factor)
    plt.show()
```

```python
# Attrition and Compensation
compensation_factors = ['Monthly Income', 'Stock Option Level']
for factor in compensation_factors:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df, x='Attrition', y=factor)
    plt.title(f'Attrition vs {factor}')
    plt.xlabel('Attrition')
    plt.ylabel(factor)
    plt.show()
```

```python
# Attrition and Training/Development
training_factors = ['Training Times Last Year', 'Education']
for factor in training_factors:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=factor, hue='Attrition')
    plt.title(f'Attrition by {factor}')
    plt.xlabel(factor)
    plt.ylabel('Count')
    plt.show()
```

```python
numerical_features  = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features  = [feature for feature in df.columns if df[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numerical_features ), numerical_features ))
print('\nWe have {} categorical features : {}'.format(len(categorical_features ), categorical_features ))
```

```python
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[feature], kde=True, bins=20, color='skyblue')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

```python
for feature in categorical_features:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=feature, palette='Set2')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Attrition by categorical features
for feature in categorical_features:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=feature, hue='Attrition', palette='pastel')
    plt.title(f'Attrition by {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.legend(title='Attrition', loc='upper right')
    plt.show()
```
